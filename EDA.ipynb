```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

def add_geo_clusters_per_city(df, lat_col='latitude', lon_col='longitude', city_col='cityname'):
    df = df.copy()
    df['geo_cluster'] = np.nan

    for city, group in df.groupby(city_col):
        coords = group[[lat_col, lon_col]].dropna()
        n = len(coords)

        if n < 100:
            continue  # Not enough data to cluster

        # Heuristic: choose number of clusters based on number of points
        # Rule of thumb: sqrt(n / 2), clamped between 2 and 10
        k = max(2, min(10, int(np.sqrt(n / 2))))

        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(coords)

        # Assign cluster labels back to the corresponding indices in original df
        df.loc[coords.index, 'geo_cluster'] = [
            f"{city}_cluster{label}" for label in cluster_labels
        ]

    return df
```

```python
city_cluster_counts = df.groupby("cityname")["geo_cluster"].count()

# Optional: filter to only cities with >=10 listings
city_cluster_counts = city_cluster_counts[city_cluster_counts >= 100]

# Sort by descending count
city_cluster_counts = city_cluster_counts.sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 6))
city_cluster_counts.plot(kind='bar', color='orange')
plt.ylabel("Number of Listings with Cluster Labels")
plt.title("Number of Geoclustered Listings per City")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
